```python
# by: Bastián Astudillo Fica
reservas1= pd.read_excel("clientes_.xlsx")
reservas1.head()
```

```python
# by: Bastián Astudillo Fica
reservas1=reservas1.sort_values('date_booking_operation',ascending=True)
```

```python
# by: Bastián Astudillo Fica
reservas = reservas1[~reservas1['status_booking'].isin(['Cancelado', 'Reemplazado', 'Pendiente', 'En revisión'])] 
```

```python
# by: Bastián Astudillo Fica
reservas["Coordenadas"] = reservas["Latitude"].map(str) + "," + reservas["Length"].map(str)
```

```python
# by: Bastián Astudillo Fica
reservas.sort_values('date_booking_operation',ascending=True).shape
```

```python
# by: Bastián Astudillo Fica
#pd.options.display.max_rows = None
geolocation=reservas[reservas['Latitude']!=0]
```

```python
# by: Bastián Astudillo Fica
geolocation
```

```python
# by: Bastián Astudillo Fica
    from geopy.exc import GeocoderTimedOut, GeocoderUnavailable

    def get_region(x):
        try:
            location = geolocator.reverse(x)
            if location is not None:
                return location.raw['address']['state']
            else:
                return None
        except (GeocoderTimedOut, GeocoderUnavailable):
            return None

# 25 minutes 
geolocation['Región'] = geolocation['Coordenadas'].apply(get_region)
```

```python
# by: Bastián Astudillo Fica
geolocation['Región'] = geolocation['Región'].fillna('0')
geolocation['Región'] = geolocation['Región'].replace('0', 'Los Angeles')
geolocation['Coordenadas'] = geolocation['Coordenadas'].replace('34.05223,-11824368.0', '34.05223,-11824368')
```

```python
# by: Bastián Astudillo Fica
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows

# Ruta del archivo Excel existente
file_path = 'clientes_.xlsx'

# Cargar el archivo Excel
workbook = openpyxl.load_workbook(file_path)

# Crear una nueva hoja en el libro de Excel
new_sheet_name = 'Geo'
new_sheet = workbook.create_sheet(title=new_sheet_name)

df = geolocation

# Convertir el DataFrame en filas para agregarlo a la hoja
rows = list(dataframe_to_rows(df, index=False, header=True))

# Agregar las filas al final de la hoja
for row in rows:
    new_sheet.append(row)

# Guardar los cambios en el archivo Excel
workbook.save(file_path)
```
